```python
def TBM(x):
    '''
    converts columns that have mix of $ Trillion, Billions and Millions to Billions
    '''
    x=str(x)
    print(x)
    try:
        num = float(re.findall(r'(.+)([BTM])',x)[0][0])
        scale = re.findall(r'(.+)([BTM])',x)[0][1]
        if scale == "B":
            return num
        elif scale == "T":
            return num*1000
        elif scale == "M":
            return num/1000
    except:
        return x
```

```python
df['op_cf'] = df['op_cf'].apply(TBM)
```
