```python
def split(x):
    x=str(x)
    print(x)
    try:
        num = float(re.findall(r'(.+)([BTM])',x)[0][0])
        scale = re.findall(r'(.+)([BTM])',x)[0][1]
        if scale == "B":
            return num
        elif scale == "T":
            return num*1000
        elif scale == "M":
            return num/1000
    except:
        return x
```
