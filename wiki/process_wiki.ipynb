```python
import pandas as pd
from datetime import datetime
```

```python
data = pd.read_csv('spx_final.csv')
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.rename({'symbol': 'ticker'}, axis=1, inplace=True)
```

```python
df.set_index('ticker')
```

```python
df.to_csv('wiki.csv')
```

```python
def date_conv(x):
    try:
        return datetime.strptime(x, '%d/%m/%Y')
    except:
        return x
```

```python
df['dt_add'] = df['dt_add'].apply(date_conv)
```
