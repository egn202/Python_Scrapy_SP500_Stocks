```python
import re
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('yhoo_profile2.csv')
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df['name_tk2'] = df['name_tk'].astype(str).apply(lambda x: re.findall(r"(.+) \((\w+)",x))
```

```python
df['ticker'] = df['name_tk2'].apply(lambda x: x[0][1])
```

```python
df['govscore2'] = df['govscore'].astype(str).apply(lambda x: re.findall(r"(\d+)",x))
```

```python
df['audit'] = df['govscore2'].apply(lambda x: x[0])
```

```python
df['board'] = df['govscore2'].apply(lambda x: x[1])
```

```python
df['sh_holdr_rgts'] = df['govscore2'].apply(lambda x: x[2])
```

```python
df['comp'] = df['govscore2'].apply(lambda x: x[3])
```

```python
df['employees'] = df.employees.str.replace(',','').astype('float')
```

```python
def pay_convert(x):
    '''
    converts columns that have mix of $ Trillion, Billions and Millions to Billions
    '''
    x=str(x)
    print(x)
    try:
        num = float(re.findall(r'(.+)([kM])',x)[0][0])
        scale = re.findall(r'(.+)([kM])',x)[0][1]
        if scale == "M":
            return num
        elif scale == "k":
            return num/1000
    except:
        return x
```

```python
df['exec1_pay'] = df['exec1_pay'].apply(pay_convert)
df['exec2_pay'] = df['exec2_pay'].apply(pay_convert)
df['exec3_pay'] = df['exec3_pay'].apply(pay_convert)
```

```python
df.drop(columns=['name_tk2', 'govscore2','name_tk'], inplace=True)
```

```python
df.set_index('ticker')
```

```python
df.to_csv('profile.csv')
```
