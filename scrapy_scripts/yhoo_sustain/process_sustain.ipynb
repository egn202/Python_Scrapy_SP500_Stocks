```python
import pandas as pd
import re
```

```python
data = pd.read_csv('yhoo_sustain.csv')
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df['name_tk2'] = df['name_tk'].astype(str).apply(lambda x: re.findall(r"(.+) \((\w+)",x))
df['ticker'] = df['name_tk2'].apply(lambda x: x[0][1])
df.drop(columns=['name_tk2','name_tk'], inplace=True)
df.set_index('ticker')
df.to_csv('sustain.csv')
```
